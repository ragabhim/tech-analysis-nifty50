```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://query1.finance.yahoo.com/v7/finance/download/%5ENSEI?period1=1557468002&period2=1589090402&interval=1d&events=history'
input_file=pd.read_csv(url, header=0,  encoding = "ISO-8859-1", parse_dates=True, index_col='Date').copy()## header=0 takes the first line as column headers
pd.options.display.max_columns=None
pd.options.display.max_rows=None
input_file= input_file.dropna(how='any') #to drop NaN rows and columns
input_file
```

Create a column, named 'Indicator' (of equal data points and in the same data frame)
The column will show 0 & 1 only.
Condition to return 1: On that particular Closing price of the day, SMA (X) crosses SMA(X+N) and closes at a higher value than SMA (X+N), where X & N > 0. Note: after the cross-over has happened, for other dates, even if SMA(X)  > SMA(X+N), then it should return 0. Only at the cross-over dates, it should return 1.
Condition to return 0: Otherwise
The return value should be in 0 & 1. Boolean and string wouldn't be counted.

```python
input_file['SMA50']= input_file['Close'].rolling(50).mean()
input_file['SMA10']=input_file['Close'].rolling(10).mean()
input_file['SMA50'].plot(figsize=(17,8), legend = True, color='black')
input_file['SMA10'].plot(figsize=(17,8), legend = True, color='blue')
plt.grid(True)
input_file.count()
```

```python
input_file['prev_50']=input_file['SMA50'].shift(1)
input_file['prev_10']=input_file['SMA10'].shift(1)
input_file['pos_indicator']=(input_file['SMA10']>input_file['SMA50']) & (input_file['prev_50']>=input_file['prev_10'])
input_file['neg_indicator']=(input_file['SMA10']<input_file['SMA50']) & (input_file['prev_50']<=input_file['prev_10'])
input_file['Indicator']=[1 if input_file.loc[ei,'pos_indicator']|input_file.loc[ei,'neg_indicator'] 
                        else 0 for ei in input_file.index]

crossover=input_file['Indicator']==1
input_file[crossover]
```

Create, a new column 'Mean_Return_NXT_10D', in the same data frame (same index and length), which will return the mean return of the stock for the next 10 days (excluding the closing price of that particular day). The 10 days are 10 stock days and not calendar days. These will iterate for index minus 10 days. To avoid NaN, slice your data frame to drop out the last 10 days

```python
input_file['Sum']=0
for j in range(1,11):
    input_file['Sum']+=input_file['Close'].shift(-j)-input_file['Close']
input_file['Mean_Return_NXT_10D']=abs(input_file['Sum']/(input_file['Close']*10))
input_file.tail()
```

```python
keep_df=input_file[['Close','Mean_Return_NXT_10D','Indicator']] 
#slices the DataFrame: to drop last 10 columns
final_df=keep_df[:-10].round({'Close':2,'Mean_Return_NXT_10D':3})
final_df
```
