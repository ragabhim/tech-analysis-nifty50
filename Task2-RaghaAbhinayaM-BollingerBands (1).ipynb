```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url='https://query1.finance.yahoo.com/v7/finance/download/%5ENSEI?period1=1559692800&period2=1591315200&interval=1d&events=history'
df=pd.read_csv(url, header=0,  encoding = "ISO-8859-1", parse_dates=True, index_col='Date').copy()## header=0 takes the first line as column headers
pd.options.display.max_columns=None
pd.options.display.max_rows=None
df= df.dropna(how='any') #to drop NaN rows and columns
df
```

```python
df['SMA20']= df['Close'].rolling(20).mean()
df['StdDev']= df['Close'].rolling(20).std()
df['UpperBand']= df['SMA20']+2*df['StdDev']
df['LowerBand']=df['SMA20']-2*df['StdDev']
df['SMA20'].plot(figsize=(17,8), legend = True, color='black')
df['Close'].plot(figsize=(17,8), legend = True, color='blue')
df['UpperBand'].plot(figsize=(17,8), legend = True, color='pink')
df['LowerBand'].plot(figsize=(17,8), legend = True, color='pink')
plt.grid(True)
df.count()
```

```python
df['Buy']=(df['Close']>df['UpperBand'])
df['Sell']=df['Close']<df['LowerBand']
df['Indicator']=[1 if df.loc[ei,'Buy']|df.loc[ei,'Sell'] else 0 for ei in df.index ]
df['BuySignal']=[5000 if df.loc[ei,'Buy'] else 0 for ei in df.index]
df['SellSignal']=[-5000 if df.loc[ei,'Sell'] else 0 for ei in df.index]
```

```python
df['Close'].plot(figsize=(17,8), legend = True, color='blue')
df['UpperBand'].plot(figsize=(17,8), legend = True, color='pink')
df['LowerBand'].plot(figsize=(17,8), legend = True, color='pink')
df['BuySignal'].plot(figsize=(17,8),legend=True, color='green')
df['SellSignal'].plot(figsize=(17,8),legend=True, color='red')
plt.grid(True)
```

```python
df['Sum']=0
for j in range(1,11):
    df['Sum']+=df['Close'].shift(-j)-df['Close']
df['Mean_Return_NXT_10D']=abs(df['Sum']/(df['Close']*10))
df.tail(15)
```

```python
keep_df=df[['Close','Mean_Return_NXT_10D','Indicator']] 
#slices the DataFrame: to drop last 10 columns
final_df=keep_df[:-10].round({'Close':2,'Mean_Return_NXT_10D':3})
final_df
```
